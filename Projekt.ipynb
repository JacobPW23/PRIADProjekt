## Sleep Health and Lifestyle Dataset

```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
path = kagglehub.dataset_download("uom190346a/sleep-health-and-lifestyle-dataset")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
column_names = df.columns
print(column_names)
df["Age"].hist(bins = 100)
```

Statystyki dotyczÄ…ce stresu

```python
stress = df['Stress Level']
print("Max stress level:", stress.max())
print("Min stress level:", stress.min())
print("Median stress level:", stress.median())
print("Mean stress level:", round(stress.mean(), 2))
df["Stress Level"].value_counts().plot(kind='bar')
```
